## 응용 피처 생성

```python
def custom_feature(df):
    df['음성사서함이용평균'] = df['음성사서함이용'] / df['가입일']
    df['음성사서함이용'] =  np.log(df['음성사서함이용'] + 1)
    df['총통화시간'] = df['주간통화시간'] + df['저녁통화시간'] + df['밤통화시간']
    df['총통화요금'] = df['주간통화요금'] + df['저녁통화요금'] + df['밤통화요금']
    df['총통화횟수'] = df['주간통화횟수'] + df['저녁통화횟수'] + df['밤통화횟수']
    for t in ['주간', '저녁', '밤']:
        df[f'1회평균{t}통화시간'] = df[f'{t}통화시간'] / df[f'{t}통화횟수']
        df[f'1회평균{t}통화요금'] = df[f'{t}통화요금'] / df[f'{t}통화횟수']
    df['1회평균통화시간'] = df['총통화시간'] / df['총통화횟수']
    df['1회평균통화요금'] = df['총통화요금'] / df['총통화횟수']
    df['상담전화건수평균'] = df['상담전화건수'] / df['가입일']
    return df
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler


if CFG['S_METHOD'] == "standard":
    scaler = StandardScaler()
elif CFG['S_METHOD'] == "minmax":
    scaler = MinMaxScaler()
else:
    print("Parameter Error!")

scaled_data = scaler.fit_transform(train_df_new.iloc[:, 1:-1])
train_df_new.iloc[:, 1:-1] = scaled_data
```

```python
from autogluon.tabular.configs.hyperparameter_configs import get_hyperparameter_config

gbm_hyperparameters = {'GBM' : get_hyperparameter_config('default')['GBM'][0]}

def evaluate(data, eval_data=False):
    predictor = TabularPredictor(label='전화해지여부', problem_type='binary', eval_metric='f1_macro', verbosity=0)
    predictor.fit(data, hyperparameters=gbm_hyperparameters, fit_weighted_ensemble=False)
    if isinstance(eval_data, pd.DataFrame):
        lb = predictor.leaderboard(eval_data, silent=True)
        score_val = lb.loc[lb.index[-1], 'score_val']
        score_test = lb.loc[lb.index[-1], 'score_test']
    else:
        lb = predictor.leaderboard(silent=True)
        score_val = lb.loc[lb.index[-1], 'score_val']
        score_test = -1
    return score_val, score_test
```

```python
def oneshot(df, feat_imp, eval_df=False):
    container = []
    for i in range(1, len(feat_imp)+1):
        score_eval, score_test = evaluate(df[feat_imp.iloc[:i].index.tolist()+['전화해지여부']], eval_df)
        temp = np.array([i, feat_imp.index[i-1], score_eval, score_test])
        container.append(temp)                 
        print(f"Feature Num: {temp[0]}", f"Added Feature: {temp[1]}", f"Score_Eval: {temp[2]}", f"Score_Test: {temp[3]}")
    history = pd.DataFrame(data=container, columns=['Iter', 'Added Feature', 'Score_Eval', 'Score_Test'])
    history['Iter'] = history['Iter'].astype(int)
    history['Score_Eval'] = history['Score_Eval'].astype(float)
    history['Score_Test'] = history['Score_Test'].astype(float)
    return history
```

def f1_macro(pred, true):
    return metrics.f1_score(true, pred, average="macro")

def evaluate(x_data, y_data):
    model = LGBMClassifier(objective='binary',  num_iterations=10**5)
    x_train, x_val, y_train, y_val = train_test_split(x_data, y_data, random_state=0)
    model.fit(x_train, y_train, eval_set=[(x_val, y_val)], early_stopping_rounds=100, verbose=False)
    val_pred = model.predict(x_val)
    score = f1_macro(val_pred, y_val)
    return score

def rfe(x_data, y_data, method, ratio=0.9, min_feats=2):
    feats = x_data.columns.tolist()[1:-1]
    archive = pd.DataFrame(columns=['model', 'n_feats', 'feats', 'score'])
    while True:
        model = LGBMClassifier(objective='binary', num_iterations=10**5)
        x_train, x_val, y_train, y_val = train_test_split(x_data[feats], y_data, random_state=0)
        model.fit(x_train, y_train, eval_set=[(x_val, y_val)], early_stopping_rounds=100, verbose=False)
        val_pred = model.predict(x_val)
        score = f1_macro(val_pred, y_val)
        n_feats = len(feats)
        print(n_feats, score)
        archive = archive.append({'model': model, 'n_feats': n_feats, 'feats': feats, 'score': score}, ignore_index=True)
        if method == 'basic':
            feat_imp = pd.Series(model.feature_importances_, index=feats).sort_values(ascending=False)
        elif method == 'perm':
            perm = PermutationImportance(model, random_state=0).fit(x_val, y_val)
            feat_imp = pd.Series(perm.feature_importances_, index=feats).sort_values(ascending=False)
        elif method == 'shap':
            explainer = shap.TreeExplainer(model)
            shap_values = explainer.shap_values(x_data[feats])
            feat_imp = pd.Series(np.abs(shap_values).mean(axis=0), index=feats).sort_values(ascending=False)
        next_n_feats = int(n_feats * ratio)
        if next_n_feats < min_feats:
            break
        else:
            feats = feat_imp.iloc[:next_n_feats].index.tolist()
    return archive
